```python
import pandas as pd

# 파일 경로 지정
file_path = r"C:\Users\Minji Kang\Documents\GitHub\network_reliability\Transparency 23-24 P01.csv"

# CSV 파일 읽기 (인코딩 지정)
df = pd.read_csv(file_path, encoding="cp949")  # 또는 encoding="euc-kr"

# 특정 열 "INCIDENT_NUMBER"의 고유 값별 등장 횟수 계산
value_counts = df["INCIDENT_NUMBER"].value_counts()

# Pandas 출력 옵션 설정 (전체 데이터 표시)
pd.set_option("display.max_rows", None)  # 모든 행 출력
pd.set_option("display.max_columns", None)  # 모든 열 출력
pd.set_option("display.expand_frame_repr", False)  # 가로로 긴 데이터도 한 줄로 표시

# 결과 출력
print(value_counts)
```
